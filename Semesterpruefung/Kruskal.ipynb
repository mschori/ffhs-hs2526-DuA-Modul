## 9Ô∏è‚É£ Python-Beispiel (Kruskal mit Union-Find)

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, a, b):
        ra, rb = self.find(a), self.find(b)
        if ra != rb:
            self.parent[rb] = ra
            return True
        return False


def kruskal(edges, n):
    uf = UnionFind(n)
    mst = []

    edges.sort(key=lambda e: e[2])  # (u, v, gewicht)

    for u, v, w in edges:
        if uf.union(u, v):
            mst.append((u, v, w))

    return mst
```
