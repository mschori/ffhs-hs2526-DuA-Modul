Selection-Sort sucht in jedem Durchlauf das **kleinste Element im unsortierten Teil** der Liste
und **tauscht es an die richtige Position** am Anfang.

- Die Liste wird in zwei Teile gedacht:
  - **links:** bereits sortiert
  - **rechts:** noch unsortiert
- In jedem Schritt w√§chst der sortierte Teil um **ein Element**

## 5Ô∏è‚É£ Warum Selection-Sort nicht stabil ist

Beispiel:

## 7Ô∏è‚É£ Python-Implementierung

````python
def selection_sort(arr):
    n = len(arr)

    for i in range(n - 1):
        min_index = i

        # Minimum im unsortierten Teil suchen
        for j in range(i + 1, n):
            if arr[j] < arr[min_index]:
                min_index = j

        # Tauschen, falls n√∂tig
        if min_index != i:
            arr[i], arr[min_index] = arr[min_index], arr[i]

    return arr
````
